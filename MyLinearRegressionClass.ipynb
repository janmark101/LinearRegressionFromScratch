```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class LinearRegression():
    def __init__(self):
        self.w = 0
        self.b = 0
        self.epochs = 10000
        self.a = 0.0005
    
    def fit(self,x,y):
        for i in range(self.epochs):
            self.w, self.b = self.gradient_descent(x,y)

    def cost_function(self,x,y):
        cost_func = 0
        temp = 0
        m = x.shape[0]
        for i in range(m):
            temp += ((self.w*x[i] + self.b) - y[i]) **2
        cost_func = 1/(2*m) * temp
        return cost_func

    def gradient_descent(self,x,y):
        m = x.shape[0]
        cost_w = 0
        cost_b = 0
        for i in range(m):
            cost_w += ((self.w*x[i] + self.b) - y[i]) * x[i]
            cost_b += ((self.w*x[i] + self.b) - y[i])
        cost_w = cost_w / m
        cost_b = cost_b /m
    
        temp_w = self.w - self.a * cost_w
        temp_b = self.b - self.a * cost_b
        
        return temp_w,temp_b

    def predict(self,x):
        return self.w * x + self.b
```

```python
data = pd.read_csv('test.csv')
```

```python
plt.scatter(data.x,data.y,color='black')
```

```python
from sklearn.model_selection import train_test_split

x = np.array(data.x)
y = np.array(data.y)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
plt.scatter(x,y,color='black')
plt.plot(x_test,y_pred,color='red')
plt.show()
```
